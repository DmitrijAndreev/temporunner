```python
import pandas as pd

from astropy import units as u
from astropy.coordinates import Angle
```

```python
common_config = pd.read_csv('common_config_1112.csv', dtype=str)
```

```python
common_config
```

```python
for idx, row in common_config.iterrows():
    ra_range = Angle(row[3])
    dec_range = Angle(row[4])
    step_ra = Angle(row[5])
    step_dec = Angle(row[6])
    
    numbers_ra = int(round(((2*ra_range)/step_ra).value))
    numbers_dec = int(round(((2*dec_range)/step_dec).value))
    
    print(
        f'STAGE {idx + 1}\n'
        f'RA iteration = {numbers_ra}\n'
        f'DEC iteration = {numbers_dec}\n'
        f'freq iteration = {row[2]}\n'
        f'FULL iteration = {numbers_ra * numbers_dec * int(row[2])}\n'
        f'#####################\n'
    )
    
    
```

```python
for idx, row in common_config.iterrows():
    with open(f'config {row[0]} {idx + 1}.py', 'w') as file:
        content = (
            f'# STAGE {idx + 1} - {row[1]}({row[2]}); {row[3]}({row[5]}); {row[4]}({row[6]});\n'
            f'\n'
            f'# Setup name of pulsar and working directory\n'
            f'name_pulsar = "{row[0]}"\n'
            f'work_dir = "{row[9]}"\n'
            f'\n'
            f'# Setup range of brute force frequency\n'
            f'freq_range = {row[2]}\n'
            f'\n'
            f'# Setup range of brute force coordinates\n'
            f'ra_range = "{row[3]}"\n'
            f'dec_range = "{row[4]}"\n'
            f'\n'
            f'# Setup step for brute force coordinates\n'
            f'step_ra = "{row[5]}"\n'
            f'step_dec = "{row[6]}"\n'
            f'\n'
            f'# on(1)/off(0) fitting params'
            f'fit_freq = {row[7]}\n'
            f'fit_coords = {row[8]}\n'
            
        )
        
        file.write(content)
```
